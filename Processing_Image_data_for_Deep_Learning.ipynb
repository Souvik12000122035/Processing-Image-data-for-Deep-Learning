```python
import cv2
# Sketch (Pencil Sketch Style)
img = cv2.imread('Taj-Mahal.jpg')
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
inv_gray = 255 - gray_img
blur_img = cv2.GaussianBlur(inv_gray, (21, 21), 0)
sketch_img = cv2.divide(gray_img, 255 - blur_img, scale=256)

cv2.imwrite('taj_sketch.jpg', sketch_img)
```

```python
plt.imshow(sketch_img, cmap='gray')
plt.axis('off')
plt.show()
```

```python
import matplotlib.pyplot as plt
# Cartoon effect
edges = cv2.adaptiveThreshold(gray_img, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 9, 9)
color_img = cv2.bilateralFilter(img, 9, 300, 300)
cartoon_img = cv2.bitwise_and(color_img, color_img, mask=edges)

cv2.imwrite('taj_cartoon.jpg', cartoon_img)
plt.imshow(cartoon_img)
plt.axis('off')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

img = cv2.imread('Taj-Mahal.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Create a sepia filter matrix
sepia_filter = np.array([[0.393, 0.769, 0.189],
                         [0.349, 0.686, 0.168],
                         [0.272, 0.534, 0.131]])

# Apply the sepia filter using dot product
sepia_img = np.dot(img.reshape(-1, 3), sepia_filter.T)

# Ensure the pixel values are within the valid range [0, 255]
sepia_img = np.clip(sepia_img, 0, 255).astype(np.uint8)

# Reshape the image back to its original dimensions
sepia_img = sepia_img.reshape(img.shape)

cv2.imwrite('taj_sepia.jpg', sepia_img)
plt.imshow(sepia_img)
plt.axis('off')
plt.show()
```

```python
#Inverted Image (Negative Effect)

import matplotlib.pyplot as plt
# Invert the image
inverted_img = 255 - img

cv2.imwrite('taj_inverted.jpg', inverted_img)
plt.imshow(inverted_img)
plt.axis('off')
plt.show()
```
